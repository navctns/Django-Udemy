```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from pandas.tools.plotting import table
from pandas import Series,DataFrame

cost_df = pd.read_table('../../data_work/bulk_cost.csv')
cost_df = cost_df[['username','date','network','country','cost']]

count_df = pd.read_table('../../data_work/bulk_count.csv')
count_df['count'] = count_df['sent']
count_df = count_df[['username','date','network','country','count']]

bulk_df = cost_df.merge(count_df, how='inner', on=['username','date','network', 'country'])

bulk_df["month_num"] =  [ i[5:7] for i in bulk_df["date"]]
bulk_df['year'] = [ i[0:4] for i in bulk_df['date']]
cost_df
```

```python
def country (country, year):
    bulk_country = bulk_df[bulk_df['country'] == country]
    bulk_country = bulk_country[bulk_country['year'] == year]
    
    def month (row):
       if row['month_num'] == '01' :
          return 'January'
       if row['month_num'] == '02':
          return 'February'
       if row['month_num'] == '03' :
          return 'March'
       if row['month_num'] == '04' :
          return 'April'
       if row['month_num'] == '05':
          return 'May'
       if row['month_num'] == '06' :
          return 'June'
       if row['month_num'] == '07' :
          return 'July'
       if row['month_num'] == '08':
          return 'August'
       if row['month_num'] == '09' :
          return 'September'
       if row['month_num'] == '10' :
          return 'October'
       if row['month_num'] == '11':
          return 'November'
       if row['month_num'] == '12' :
          return 'December'
       return 'Other'
    
    %matplotlib inline
    bulk_country['month'] = bulk_country.apply(lambda row: month(row), axis=1)
    bulk_country = bulk_country[bulk_country['month'] != 'April']
    
    bulk_country['cost'] = pd.to_numeric(bulk_country['cost'])
    bulk_country['count'] = pd.to_numeric(bulk_country['count'])
    
    bulk_country_monthly_count = bulk_country[['month', 'month_num', 'count']].groupby(['month', 'month_num'], as_index=False).sum().sort_values(by="month_num", axis=0, ascending=True, inplace=False)
    bulk_plot_count = bulk_country_monthly_count[['month', 'count']]
    
    bulk_country_monthly_cost = bulk_country[['month', 'month_num', 'cost']].groupby(['month', 'month_num'], as_index=False).sum().sort_values(by="month_num", axis=0, ascending=True, inplace=False)
    bulk_country_monthly_cost['cost'] = bulk_country_monthly_cost['cost'].round(2)
    bulk_plot_cost = bulk_country_monthly_cost[['month', 'cost']]
    
    bulk_plot = bulk_plot_cost.merge(bulk_plot_count, how='inner', on='month').set_index('month').transpose()
    
    f, (ax) = plt.subplots(figsize=(12, 6), sharex=True)
    ax1 = ax.twinx()
    
    sns.barplot(x='month', y='cost', data=bulk_country_monthly_cost, ax=ax, color='g')
    
    sns.pointplot(x='month', y='count', data=bulk_country_monthly_count, table=True, ax=ax1, color='b')
    ax.get_xaxis().set_visible(False)
    table(ax, bulk_plot)
    ax.set(xlabel='Month', ylabel='Cost(USD)')
    ax1.set(ylabel='Count(messages)')
    sns.plt.title(country + ' Bulk SMS ' + year, y=1.05)
    
    count_patch = mpatches.Patch(color='blue', label='Count')
    cost_patch = mpatches.Patch(color='green', label='Cost')
    plt.legend(handles=[cost_patch, count_patch], loc='upper left', bbox_to_anchor=(0.0,1.1))
    
    ax.get_figure().savefig('../../reports/live/bulksms/' + country + '/bulksms_overall.png')

    top_10 = bulk_country[['username','cost', 'count']].groupby(['username'], as_index=False).sum().sort_values(by="cost", axis=0, ascending=False, inplace=False).head(10)
    top_10['cost'] = top_10['cost'].round(2)
    top_list = top_10['username'].values.tolist()
    index = range(1,len(top_list)+1)
    top_list = DataFrame(top_list, index=index, columns=['username'])
    top_list = [tuple(x) for x in top_list.to_records(index=True)]
    
    top_10_plot = top_10.set_index('username').transpose()
    
    f, (ax2) = plt.subplots(figsize=(12, 6), sharex=True)
    ax3 = ax2.twinx()
    
    sns.barplot(x='username', y='cost', data=top_10, ax=ax2, color='g')
    
    sns.pointplot(x='username', y='count', data=top_10, table=True, ax=ax3, color='b')
    ax2.get_xaxis().set_visible(False)
    table(ax2, top_10_plot)
    
    ax2.set(xlabel='Username', ylabel='Cost(USD)')
    ax3.set(ylabel='Count(messages)')
    
    sns.plt.title(country + ' Top 10 Users', y=1.05)
    
    count_patch = mpatches.Patch(color='blue', label='Count')
    cost_patch = mpatches.Patch(color='green', label='Cost')
    plt.legend(handles=[cost_patch, count_patch], loc='upper left', bbox_to_anchor=(0.0,1.1))
    
    ax2.get_figure().savefig('../../reports/live/bulksms/' + country + '/bulksms_topten.png')
    
    
    def user(num, username): 
        user_monthly = bulk_country[bulk_country['username'] == username][['month','month_num','cost', 'count']].groupby(['month','month_num'], as_index=False).sum().sort_values(by="month_num", axis=0, ascending=True, inplace=False)
        user_monthly['cost'] = user_monthly['cost'].round(2)
        user_plot = user_monthly[['month','cost','count']].set_index('month').transpose()
        
        f, (ax4) = plt.subplots(figsize=(12, 6), sharex=True)
        ax5 = ax4.twinx()
        
        sns.barplot(x='month', y='cost', data=user_monthly, ax=ax4, color='g')
        
        sns.pointplot(x='month', y='count', data=user_monthly, table=True, color='b', ax=ax5)
        
        ax4.get_xaxis().set_visible(False)
        table(ax4, user_plot)
        
        ax4.set(xlabel='Month', ylabel='Cost(USD)')
        ax5.set(ylabel='Count(messages)')
        
        sns.plt.title(username, y=1.05)
        
        count_patch = mpatches.Patch(color='blue', label='Count')
        cost_patch = mpatches.Patch(color='green', label='Cost')
        plt.legend(handles=[cost_patch, count_patch], loc='upper left', bbox_to_anchor=(0.0,1.1))
        
        ax4.get_figure().savefig('../../reports/live/bulksms/' + country + '/user_'+ str(num) + '.png')
        
    for username in top_list:
        user(username[0], username[1])
        
    return 'success'
```

```python
countries = ['Rwanda','Malawi','Nigeria','Kenya','Uganda','Tanzania']

for c in countries:
    country(c, '2017')
```
